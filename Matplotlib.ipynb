MatplotLib:-Useful for making plot

```python
#importing matplotlib library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(0,10,100) # here is value from 0 to 10 with 100 elements
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
# Adding labels in x-axis and y-axis
#sine wave
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value ')
plt.title('sine wave ')
plt.show()
```

```python
#cos wave
x=np.linspace(0,10,100) # here is value from 0 to 10 with 100 elements
z=np.cos(x)
plt.plot(x,z ,'r--')
plt.xlabel('angle')
plt.ylabel('cos value ')
plt.title('cos wave ')
plt.show()
```

```python
#parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
#plot x and y with  red color and with + symbol
plt.plot(x,y, 'r+')
plt.show()
```

```python
#plot x and y with  green color and with . symbol
plt.plot(x,y, 'g.')
plt.show()
```

Bar plot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])  # Here (0,0 ) represent the coordinate ans (1,1) repn height and width of coordinate
language=['engish','hindi','telugu','marathi','Bhojpuri']
people=[100,50,150,40,70]
ax.bar(language,people)
plt.xlabel('LANGUAGES')
plt.ylabel(' NO. OF PEOPLE')
plt.show()
```

Pie- Chart

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
language=['engish','hindi','telugu','marathi','Bhojpuri']
people=[100,50,150,40,70]
ax.pie(people,labels=language,autopct='%1.1f')
plt.show()
```

Scatter plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

3D Scatter plot

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
