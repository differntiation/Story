Label Encoding :Converting the labels into numeric form

```python
#Importing the dependency
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of diabetes dataset

```python
#loading the breast_cancer data from csv file to pandas dataframe
cancer_data=pd.read_csv('/content/breast_cancer.csv')
```

```python
cancer_data.head()
```

Above in diagonosis column two stage are written

1.   M-Malign stage i.e advanced stage
2.   balign Stage i.e initial stage and give treatment early on



```python
#Finding the count of diffrent labels
cancer_data['diagnosis'].value_counts()
```

```python
#load the label ecoder function
label_encode=LabelEncoder()
```

```python
labels=label_encode.fit_transform(cancer_data.diagnosis)
```

```python
#Appending the labels to the dataframe
cancer_data['target']=labels
```

```python
cancer_data.head()
```

Above we can see that New "target" column add in which "Malignant->1 and Benign->0"  since  number ois give alphabetical order B comes first so given 0 .now we can drop diagnosis col

```python
cancer_data['target'].value_counts() #we can compare with prev one
```

Label Encoding of Iris dataset -> contains three labels
