Imbalnced Dataset:-An dataset with an unequal class distribution, Suppose we have dataset which has 900 diabetic  patient and only 100 non-diabetic patient then it is very difficult to predict the result correctly

```python
#importing  The dependencies
import numpy as np
import pandas as pd
```

```python
#loading the dataset to pandas dataFrame
credit_card_data=pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

Above in class col we have 0->legal transaction and 1-> fraud transaction

```python
credit_card_data.tail()
```

```python
#distribution of two class
credit_card_data['Class'].value_counts()
```

Above we can see that there more than 2 lakh  data for legal transaction and only 492 for fraud transaction so this is highly imbalanced dataset

```python
#separating the legit and fraudlent
legit=credit_card_data[credit_card_data.Class==0]
fraud =credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Undersampling : in this i will take similar no. of random legit transaction same as fraud transaction

Build sample dataset containing similar distribution of legit and fraudlent transaction

No.of Fraud transaction =492

```python
legit_sample=legit.sample(n=492) # we have to take the similar no. of legit transaction
```

```python
print(legit_sample.shape)
```

Concatenate the two dataFrame first 492 is legit_smaple and then 492 is fraud sample

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0) #  here axis-0 means we are concatenating row wise and if axis =1 then we cocatenate column wise
```
